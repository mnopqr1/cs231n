**My calculation of the gradient**

The weight matrix $W$ is (in the provided implementation) a matrix with $d$ rows and $c$ columns. The $j$th column vector $W_j$ stands for the "average picture" for class $j$.

The gradient of the loss function is $\sum_{i=1}^N \frac{dL_i}{dW}$ where $L_i$ is the loss at training example $x_i$. The loss $L_i$ at training example $x_i$ is itself a sum, $\sum_{j \neq y_i} L_{i,j}$, where $L_{i,j} := \max(0, x_i \cdot (W_j - W_{y_i}) + 1)$.

For each class $j \neq y_i$, we therefore have that $\frac{dL_i}{dW_j} = \frac{dL_{i,j}}{dW_j} = x_i$ if the margin ($x_i \cdot (W_j - W_{y_i}) + 1$) for class $j$ is greater than 0, and $0$ otherwise.

For the correct class $y_i$, we see that $L_{i,j}$ contributes $-x_i \cdot W_{y_i}$ to $L_i$ for every $j$ where the margin is greater than $0$. So the contribution to $dW_{y_i}$ is $-x_i$ every time there is a margin greater than $0$.

```python
# Once you've implemented the gradient, recompute it with the code below
# and gradient check it with the function we provided for you

# Compute the loss and its gradient at W.
loss, grad = svm_loss_naive(W, X_dev, y_dev, 0.0)

# Numerically compute the gradient along several randomly chosen dimensions, and
# compare them with your analytically computed gradient. The numbers should match
# almost exactly along all dimensions.
from cs231n.gradient_check import grad_check_sparse
f = lambda w: svm_loss_naive(w, X_dev, y_dev, 0.0)[0]
grad_numerical = grad_check_sparse(f, W, grad)

# do the gradient check once again with regularization turned on
# you didn't forget the regularization gradient did you?
loss, grad = svm_loss_naive(W, X_dev, y_dev, 5e1)
f = lambda w: svm_loss_naive(w, X_dev, y_dev, 5e1)[0]
grad_numerical = grad_check_sparse(f, W, grad)
```

# Some more experiments with the hyperparameters and pictures

The best values I found were $LR = 10^{-7}$ and $RS = 2.5 * 10^4$. This gave a test set accuracy of 38.5% after 1500 iterations. More iterations than that makes the model worse. The pictures are above.

Here I show some more experiments with extreme values of the hyperparameters. 

1. If you make the regularization strength very high (40 times higher than optimal), the pictures start to look more like the actual things but the accuracy goes down (31% on the test set). I guess this is overfitting in action.
2. If you make the learning rate too high (10 times smaller than optimal), the pictures get grainy and loss starts to fluctuate during training. Accuracy is really bad (19%).
3. Perhaps the most interesting: if you make regularization strength very low ($RS = 10^3$, which is 25 times smaller than optimal) but keep learning rate the same, **and train 4 times as long**, then the test accuracy is about the same as with the optimal hyperparameters (38.2%), and validation accuracy is even higher (39.4%). However, the pictures then look like almost random noise!

```python
def experiment_with(lr, rs, num_iters=1500, verbose=True):
    print(f"learning rate: {lr}, regularization strength: {rs}")
    svm = LinearSVM()
    loss_hist = svm.train(X_train, y_train, learning_rate=lr, reg=rs,
                      num_iters=num_iters, verbose=verbose)
    y_train_pred = svm.predict(X_train)
    y_val_pred = svm.predict(X_val)
    
    train_acc = np.mean(y_train == y_train_pred)
    val_acc = np.mean(y_val == y_val_pred)
    
    print(f"training accuracy: {train_acc}, validation accuracy: {val_acc}")
    
    y_test_pred = svm.predict(X_test)
    test_accuracy = np.mean(y_test == y_test_pred)
    print('linear SVM on raw pixels final test set accuracy: %f' % test_accuracy)
    
    visualize(svm)
```

```python
experiment_with(1e-7, 1e6) # regularization too high
```

```python
experiment_with(1e-5, 1e4) # learning rate too low
```

```python
experiment_with(1e-7, 1e3, 6000) # low regularization, high number of iterations
```
